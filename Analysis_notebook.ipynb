```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as datetime

%matplotlib inline
```

```python
# Import the data from the local dekstop

df = pd.read_csv("membership_data.csv")
```

```python
df.head()
```

```python
# Convert the columns started_at and ended_at to datetime object

df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])
df["ride_length"] = df.ended_at - df.started_at
```

```python
# After the calculation the ride_length series became Timedelta object then I converted it back to datetime object by adding
# ride_length with date

date = datetime.datetime(2023, 11, 22)
df.ride_length = date + df.ride_length
df.ride_length =df.ride_length.dt.time
```
